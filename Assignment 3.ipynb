```python
shape = df.shape
print(f"The shape of the DataFrame is: {shape}")
print(f"Number of rows: {shape[0]}")
print(f"Number of columns: {shape[1]}")
```

```python
print("\nSummary of the DataFrame:")
print(df.info())

print("\nDescriptive statistics of the DataFrame:")
print(df.describe())
```

```python
print("First 5 rows of the DataFrame:")
print(df.head())
print("\nLast 3 rows of the DataFrame:")
print(df.tail(3))
```

```python
average_salary = df['Salary'].mean()
print(f"The average salary of employees is: {average_salary}")

total_bonus = df['Bonus'].sum()
print(f"The total bonus paid to all employees is: {total_bonus}")

youngest_age = df['Age'].min()
print(f"The youngest employee's age is: {youngest_age}")

highest_bonus = df['Bonus'].max()
print(f"The highest performance rating (represented by Bonus) is: {highest_bonus}")
```

```python
df_sorted = df.sort_values(by='Salary', ascending=False)
df_sorted
```

```python
def categorize_performance(rating):
    if rating >= 4.5:
        return "Excellent"
    elif rating >= 4.0:
        return "Good"
    else:
        return "Average"

df['PerformanceCategory'] = df['Bonus'].apply(categorize_performance)
df
```

```python
missing_values = df.isnull().sum()
print("\nMissing values per column using isnull():\n", missing_values)
```

```python
df = df.rename(columns={"EmployeeID": "ID"})
df.to_csv("employees.csv", index=False)
df
```

```python
experienced_employees = df[df['YearsOfExperience'] > 5]
experienced_employees
```

```python
IT_department= df[df['Department'] == 'IT']
IT_department
```

```python
df['Tax'] = df['Salary'] * 0.10
df
```
